
The failure probability is given by an exponential distribution. Thus, if "t" is the time to failure, the associated cumulative distribution function FT(t) is given as: FT(t) = 1 - exp(-2t).
You want to generate a string of 500 times to failure (e.g., t1, t2, â€¦t500) for this random process based on a Monte Carlo scheme involving generation of random numbers ri. Obtain a formula linking the ith failure time ti to the ith random number ri.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
ri=np.random.random(500)
```

```python
rv=[]
for n in ri:
    rv.append([np.log(n/2)/-2,n])
rv=np.array(rv)

plt.scatter(rv.T[0],rv.T[1])
plt.xlabel("time to failure")
plt.ylabel("random number")

```
