```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
particles=np.zeros(10_000)
moves=(-.01,0.01)
```

```python
for i in range(50):
    particles+=np.random.choice(moves,size=10_000)

plt.hist(particles)
plt.xlabel("diffusion displacement")
plt.ylabel("No. of particles")
plt.xlim(-1,1)
plt.title("Particle Diffusion after 50 time steps")
```

```python
for i in range(500):
    particles+=np.random.choice(moves,size=10_000)

plt.hist(particles)
plt.xlabel("diffusion displacement")
plt.ylabel("No. of particles")
plt.xlim(-1,1)
plt.title("Particle Diffusion after 500 time steps")
```
